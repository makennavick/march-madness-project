```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Import pandas and read the charity_data.csv from the provided cloud URL.
import pandas as pd
bracket_df = pd.read_csv('Data_cleaning/barttorvik_all_years.csv')
bracket_df.head()
```

```python
#Exclude 2025 data from CSV

#Create Copy of Data
copy_df = bracket_df.copy()

#Create New Dataframe for all years except 2025
no_2025_df = copy_df.loc[(copy_df["Year"] != 2025)]
```

```python
#Create Dummies Dataframe
dummies_df = pd.get_dummies(no_2025_df)
dummies_df.head()
```

```python
#separate Data into features and target arrays
y = dummies_df["Round Finished"]
X = dummies_df.drop(columns="Round Finished")

#Split the data into training and testing dataset\
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Create Neural network model
nn = tf.keras.models.Sequential()

#Create input Layer
nn.add(tf.keras.layers.Input(shape=[43]))

# First hidden layer
nn.add(tf.keras.layers.Dense(units=7, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=8, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```
