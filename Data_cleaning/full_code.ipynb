```python
import pandas as pd

#test how I would retreive and clean data

# Define the URL
year = 2008  
url = f"https://barttorvik.com/trank.php?year={year}"

# Read the table from the URL
tables = pd.read_html(url) 

# Inspect the first table
df = tables[0]  

# Display the DataFrame
df.head(341)
```

```python
print(df.columns)
```

```python
# Drop the first level of multiIndex columns
df.columns = df.columns.droplevel(0)

# Display updated DataFrame
print(df.head())
```

```python
new_df = df[df[("Team")].astype(str).str.contains("seed", na=False)]
new_df.head(65)
```

```python
# List of specific columns to modify
cleaned_columns = ["AdjOE", "AdjDE", "Barthag", "EFG%", "EFGD%", "TOR", "TORD", "ORB", "DRB",
                    "FTR", "FTRD", "2P%", "2P%D", "3P%", "3P%D", "3PR", "3PRD", "Adj T.", "WAB"]

# Apply transformation only to selected columns
new_df.loc[:, cleaned_columns] = new_df[cleaned_columns].apply(lambda col: col.astype(str).str.split(" ").str[0])


# Display the cleaned DataFrame
new_df.head()
```

```python
# Extract the School Name
new_df.loc[:, 'School'] = new_df['Team'].str.extract(r'(.+?)\s+\d+\sseed')

# Extract the Seed Rank
new_df.loc[:, 'Seed Rank'] = new_df['Team'].str.extract(r'(\d+\sseed)')

# Extract the Round Finished (if present)
new_df.loc[:, 'Round Finished'] = new_df['Team'].str.extract(r'\d+\sseed,\s*(.*)')


new_df.loc[:, 'Year'] = year
```

```python
new_df.head()
```

```python
#Get all years data with For Loop
```

```python
# Define the range of years
years = range(2008, 2026)

# Initialize a list to store DataFrames
df_list = []

# Loop through each year, fetch data, process it, and append to the list
for year in years:
    url2 = f"https://barttorvik.com/trank.php?year={year}&sort=&top=0&conlimit=All&venue=All&type=R#"
    try:
        # Read tables from URL
        tables2 = pd.read_html(url2)
        df2 = tables2[0] 

        # Drop the first level of MultiIndex columns if present
        if isinstance(df2.columns, pd.MultiIndex):
            df2.columns = df2.columns.droplevel(0)

        # Filter only teams with "seed" in their name
        df_filtered = df2[df2["Team"].astype(str).str.contains("seed", na=False)].copy()

        # List of specific columns to clean
        columns_to_clean = ["AdjOE", "AdjDE", "Barthag", "EFG%", "EFGD%", "TOR", "TORD", "ORB", "DRB",
                            "FTR", "FTRD", "2P%", "2P%D", "3P%", "3P%D", "3PR", "3PRD", "Adj T.", "WAB", "Rec"]

        # Extract first numeric value before space in each column
        df_filtered[columns_to_clean] = df_filtered[columns_to_clean].astype(str).apply(lambda col: col.str.split().str[0])

        # Extract School Name, Seed Rank, and Round Finished
        df_filtered["School"] = df_filtered["Team"].str.extract(r'(.+?)\s+\d+\sseed')
        df_filtered["Seed Rank"] = df_filtered["Team"].str.extract(r'(\d+\sseed)')
        df_filtered["Round Finished"] = df_filtered["Team"].str.extract(r'\d+\sseed,\s*(.*)')

        # Add Year column
        df_filtered["Year"] = year

        # Append to list
        df_list.append(df_filtered)
        print(f"Successfully processed data for {year}")

    except Exception as e:
        print(f"Failed to fetch or process data for {year}: {e}")

# Combine all DataFrames into one
df_combined = pd.concat(df_list, ignore_index=True)
```

```python
df_combined.head()
```

```python
# Clear Round Finished values for Year 2025
df_combined.loc[df_combined["Year"] == 2025, "Round Finished"] = ""
```

```python
df_combined["AdjOE"] = pd.to_numeric(df_combined["AdjOE"])
df_combined["AdjDE"] = pd.to_numeric(df_combined["AdjDE"])

df_combined["AdjNR"] = df_combined["AdjOE"] - df_combined["AdjDE"]
```

```python
# Select final columns
final_columns = ["Year", "Rk", "School", "Seed Rank", "Round Finished", "Conf", "G", "Rec", 
                 "AdjOE", "AdjDE", "AdjNR", "Barthag", "EFG%", "EFGD%", "TOR", "TORD", 
                 "ORB", "DRB", "FTR", "FTRD", "2P%", "2P%D", "3P%", "3P%D", 
                 "3PR", "3PRD", "Adj T.", "WAB"]

# Filter final DataFrame
final_df = df_combined[final_columns]

```
